## Задание 3

```python
# Окружность
x = []
y1 = []
y2 = []
R = 2
for i in range(1000*2*R+1):
    x_ = -R + i/1000
    x.append(x_)
    y1.append(math.sqrt(R**2 - x_**2))
    y2.append(-math.sqrt(R**2 - x_**2))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3,3)
plt.xlim(-4.3,4.3)
plt.grid(True)
```

```python
# Эллипс
x = []
y1 = []
y2 = []
a = 4
b = 2
for i in range(1000*2*a+1):
    x_ = -a + i/1000
    x.append(x_)
    y1.append(math.sqrt(1 - x_**2/a**2)*b)
    y2.append(-math.sqrt(1 - x_**2/a**2)*b)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3,3)
plt.xlim(-4.3,4.3)
plt.grid(True)
```

```python
# Гипербола
x = []
y1 = []
y2 = []
a = 1
b = 3
for i in range(1000*2*a):
    x_ = a + i/1000
    x.append(x_)
    y1.append(math.sqrt(x_**2/a**2-1)*b)
    y2.append(-math.sqrt(x_**2/a**2-1)*b)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3,3)
plt.xlim(-3,3)
plt.grid(True)
```

```python
# Трехмерный график двух параллельных плоскостей
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,1)
Y = np.arange(-5,5,1)
X, Y = np.meshgrid(X,Y)
Z = 2*X + 3*Y
Z2 = 2*X + 3*Y + 100
ax.plot_surface(X,Y,Z)
ax.plot_surface(X,Y,Z2)
show()
```

```python
# Трехмерный график двух поверхностей второго порядка
fig = figure()
ax = Axes3D(fig)
X = np.arange(-20,20,1)
Y = np.arange(-20,20,1)
X, Y = np.meshgrid(X,Y)
Z = np.sqrt(X**2 + Y**2)
Z2 = X**2/10 + Y**2/10
ax.plot_surface(X,Y,Z)
ax.plot_surface(X,Y,Z2)
show()
```

```python
# Перевод из полярных в декартовы 
phi = np.linspace(0, 2*np.pi, 1000)
rho = 10
x = rho * np.cos(phi)
y = rho * np.sin(phi)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
show()
```

```python
# График окружности в полярных координатах
t = np.linspace(0, np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
rho = np.sqrt(x**2 + y**2)
phi = np.arctan2(y,x)*100/np.pi
plt.polar(phi, rho)
show()
```
